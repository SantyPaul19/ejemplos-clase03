```python
import pandas as pd
```

```python
data = pd.read_csv("misdata.csv")
```

```python
data
```

```python
def obtener_inicial(d):
    cadena = d[0]
    return cadena
```

```python
def obtener_mayuscula(d):
    cadena = d.upper()
    return cadena
```

```python
data['inicial'] = data['nombre'].apply(obtener_inicial)
```

```python
data
```

```python
data['mayuscula'] = data['nombre'].apply(obtener_mayuscula)
```

```python
data
```

```python
data.to_csv("misdata2.csv")
```
